```python
csv_data ='/content/drive/MyDrive/Colab Notebooks/winrate/Final_level_data/all_levels_processed_data.csv'
# Define an empty list to store each chunk
chunks = []
all_lvls_df = pd.DataFrame()

# Load file in chunks of 100,000 rows at a time
for chunk in pd.read_csv(csv_data, chunksize=2000):
    # Process each chunk here (optional)
    # chunks.append(chunk)
    all_lvls_df = chunk
    break

# Concatenate all chunks
# data = pd.concat(chunks, axis=0)

# data.head()

all_lvls_df.head(2)
```

```python
drp_col =['In Range?','Low Range','High Range']

all_lvls_df.drop(drp_col, axis=1, inplace=True)

all_lvls_df.set_index('level', inplace=True)
```

```python
all_lvls_df.head(1)
```

```python
drp_col =['Unnamed: 0']

all_lvls_df.drop(drp_col, axis=1, inplace=True)
```

```python
all_lvls_df.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
np.random.seed(0)
train_lvls, test_lvls = train_test_split(all_lvls_df, train_size = 0.8,test_size = 0.2, random_state = 42)
```

```python
from pycaret.regression import setup, compare_models, predict_model

# Set up the PyCaret regression environment
regression_setup = setup(train_lvls, target='winrate',
                         session_id=123,
                         normalize=True, use_gpu=True)
```

```python
# Compare models and select the best
best_model = compare_models()
```

```python
from pycaret.regression import save_model

# Save the finalized model
save_model(best_model, '/content/drive/MyDrive/Colab Notebooks/winrate/model/best_model')
```

```python
# Predict on the original dataset
(predict_model(best_model, test_lvls)).head()
```

```python
from pycaret.regression import finalize_model

# Finalize the model
final_best_model = finalize_model(best_model)
```

```python
from pycaret.regression import save_model

# Save the finalized model
save_model(final_best_model, '/content/drive/MyDrive/Colab Notebooks/winrate/model/best_lvls_model')
```

```python
from pycaret.regression import load_model

# Load the saved model
final_best_model = load_model('/content/drive/MyDrive/Colab Notebooks/winrate/model/best_lvls_model')
```
