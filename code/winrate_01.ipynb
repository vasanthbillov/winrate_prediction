```python

# Load the dataset from Excel
lvls_cl_df = pd.read_excel('Lvls_1000_cleaned_df_v3_no_gameboad_data.xlsx', index_col=0)

lvls_cl_df.head()
```

```python
np.random.seed(0)
train_lvls, test_lvls = train_test_split(lvls_cl_df, train_size = 0.9,
                                                       test_size = 0.1, random_state = 42)
```

```python
# lvl=lvls_cl_df[['moves', 'score', 'frogs', 'bugs', 'r_weights', 'b_weights',
#             'g_weights', 'y_weights', 'v_weights', 'o_weights', 'w_weights','c_weights', 'h_weights', 'p_weights', 'z_weights', 't_weights']]
```

```python
from pycaret.regression import setup, compare_models

# Set up the PyCaret regression environment
regression_setup = setup(train_lvls, target='p3_winrate',
                         session_id=123,
                         normalize=True)

```

```python
#TypeError: setup() got an unexpected keyword argument 'silent'
```

```python
# Compare models and select the best
best_model = compare_models()
```

```python
from pycaret.regression import predict_model

# Predict on the original dataset
predict_model(best_model, test_lvls)
```

```python
from pycaret.regression import finalize_model

# Finalize the model
final_best_model = finalize_model(best_model)
```

```python
from pycaret.regression import save_model

# Save the finalized model
save_model(final_best_model, 'best_lvls_model')
```

```python
from pycaret.regression import load_model

# Load the saved model
final_best_model = load_model('best_lvls_model')
```

```python
from pycaret.regression import predict_model

# Predict on the original dataset
predictions = predict_model(final_best_model)

# Show the predictions
print(predictions[['p3_winrate']])
```
